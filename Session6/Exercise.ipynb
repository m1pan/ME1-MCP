```python
#TaskA

def fibonnaci(n):
    if n == 1 or n ==2:
        return 1
    sum = fibonnaci(n-1) + fibonnaci(n-2)
    return sum

print(fibonnaci(16))
```

```python
#Task2
def reverse(a):
    if len(a) > 1:
        x = a[-1] + reverse(a[:-1])
    else:
        x = a
    return x
print(reverse("desserts"))
```

```python
def permute(x):
    if len(x) == 1:
        return [x]
    perms = []

    for a in x:
        tmp = [b for b in x if b != a]
        z = permute(tmp)
        for b in z:
            perms.append([a] + b)

    return perms

print(permute([1,2,3,4]))
```

```python
def gcd(a, b):
    if b == 0:
        return a
    else:
        return gcd(b, a % b)
    
```

```python
def addFractions(N, D):
    if len(N) == 1:
        return (N[0]/D[0])
    sum = N[0]/D[0] + addFractions(N[1:], D[1:])
    return sum
print(addFractions([3,1,3,2,5,7,2,4,6,1,7], [4,5,8,16,2,4,7,2,4,3,36]))
```

```python
import matplotlib.pyplot as plt

# read data from file
with open("Temperatures.txt") as f:
    data = [float(line) for line in f]

# compute average temperature for each day
avg_temp = [sum(data[i:i+24])/24 for i in range(0, len(data), 24)]

# compute maximum and minimum temperature
max_temp = max(data)
min_temp = min(data)

# plot the results
plt.plot(avg_temp)
plt.xlabel('Day')
plt.ylabel('Average Temperature (°C)')
plt.title('Average Temperature per Day')
plt.show()

# print the results
print("Maximum temperature:", max_temp, "°C")
print("Minimum temperature:", min_temp, "°C")
print(avg_temp)
```
