```python
# write the function
def Series(N):
    a = 5
    # initialise the sum
    S = 0
    for j in range(-N,N+1):
        # initialise the sum of the internal series
        Si = 0
        # calculate the internal series in the prescribed range
        for k in range(-j,j):
            if k !=0:
                Si += (-1)**j*np.sin(a*k)/k**a
        # update the value of S
        S += Si/(100-a**2)
    return S


# prepare an array to store N and S
N = np.zeros(20)
S = np.zeros(20)
# compute S for each N
for i in range(1,21):
    N[i-1] = i
    S[i-1] = Series(int(i))
# plot the results
pl.plot(N,S)

        
        
```

```python
# from Session 8
def Transpose(A):
    N = A.shape[0]
    M = A.shape[1]
    AT = np.zeros((M,N))
    for i in range(M):
        for j in range(N):
            AT[i,j] = A[j,i]
    return AT

def Minor(A,i,j):
    N = A.shape[0]
    B = np.zeros((N-1,N-1))
    B[:i,:j] = A[:i,:j]
    B[:i,j:] = A[:i,j+1:]
    B[i:,:j] = A[i+1:,:j]
    B[i:,j:] = A[i+1:,j+1:]
    return B

def Determinant(A):
    N = A.shape[0]
    if N == 1:
        det = A[0,0]
    else:
        det = 0
        for j in range(N):
            det += (-1)**(0+j)*A[0,j]*Determinant(Minor(A,0,j))
    return det

def Adjoint(A):
    N = A.shape[0]
    Ad = np.zeros((N,N)) 
    for i in range(N):
        for j in range(N):
            Ad[i,j] = (-1)**(i+j)*Determinant(Minor(A,i,j))
            print(Minor(A,i,j))
    Ad = Transpose(Ad)    
    return Ad

def Inverse(A):
    N = A.shape[0]
    AI = np.zeros((N,N))
    d = Determinant(A)
    if d != 0:
        AI = Adjoint(A) / d
    else:
        AI = 0
    return AI
```

```python
# read the data in
Nw = 10 # size of submatrices (Nw x Nw)
Ns = 60 # number of submatrices (Ns x Ns)

# read the content of the file
f = open('Matrix.txt','r')
tmp = f.readlines()
f.close()
# organise the data into the array A of size 600 x 600
A = np.ndarray((Ns*Nw,Ns*Nw))
c = 0 # counter within the list
for i in range(Ns*Nw):
    for j in range(Ns*Nw):
        A[i,j] = float(tmp[c]) # convert into a number and put into the array
        c += 1
pl.imshow(A)
pl.show()
    
# average of submatrices
AV = np.ndarray((Ns,Ns))
# determinants of subnatrices
Det = np.ndarray((Ns,Ns))
# submatrix
Sub = np.ndarray((Nw,Nw))
# compute, for every submatrix
for i in range(Ns):
    for j in range(Ns):
        # form the submatrix
        Sub = A[i*Nw:i*Nw+Nw,j*Nw:j*Nw+Nw]
        # find the average
        AV[i,j] = sum(sum(Sub))/(Nw*Nw) # use function sum or do it yourself:
        #S = 0
        #for m in range(Nw):
        #    for n in range(Nw):
        #        S += Sub[m,n]
        #AV[i,j] = S/(Nw*Nw)
        
        # find the determinant
        Det[i,j] = np.linalg.det(Sub) # use numpy function or
        # do it yourself making use of what we did in the tutorial and reusing the same functions
        #Det[i,j] = Determinant(Sub)
        
        # establish if to compute the transpose of the submatrix
        if Det[i,j] !=0:
            # find the transpose
            SubTr = np.transpose(Sub) # use numpy function or
            # do it yourself making use of what we did in the tutorial and reusing the same functions
            #SubTr = Transpose(Sub)
            # substitute back into large matrix A
            A[i*Nw:i*Nw+Nw,j*Nw:j*Nw+Nw] = SubTr

# save results into files
fav = open('Averages.txt','w')
fdet = open('Determinants.txt','w')
for i in range(Ns):
    for j in range(Ns):
        fav.write(str(AV[i,j])+'\n')
        fdet.write(str(Det[i,j])+'\n')
fav.close()
fdet.close()

            
# plot the matrix as an image
pl.imshow(A)
```

```python
# set max number of moves
Nmax = 1000

# plot the spatial domain
Rad = 0.4
Rtheta = range(0,360)
for theta in Rtheta:
    thetar = theta * np.pi / 180
    pl.scatter(Rad*np.cos(thetar)-1,Rad*np.sin(thetar)+2,s=3,c='b')
    pl.scatter(Rad*np.cos(thetar)-2,Rad*np.sin(thetar)+1,s=3,c='b')
pl.axes().set_aspect('equal')  
bx = np.arange(-4,0.1,0.1)
by = bx + 4
pl.plot(bx,by,Linewidth=3,c='b')
bx = np.arange(-2,0.1,0.1)
by = bx + 2
pl.plot(bx,by,Linewidth=3,c='b')
pl.plot([-4,-2],[0,0],Linewidth=3,c='b')
pl.grid()

# set max number of moves
Nmax = 1000
# set the ant in the initial position
x = -3.0
y = 0.0
# move the ant
nmoves = 0
while x<0 and nmoves < Nmax:  # there is no conditon to exit or still not reached max moves
    # register a new move
    nmoves += 1
    # store the current position as previous one
    # this is needed as we want to plot the move from a to b as a line
    xp = x
    yp = y
    # compute the size of the move, dx, dy
    dx = rn.random()*0.6-0.3
    dy = rn.random()*0.6-0.3
    # do th emove: update the coordinates of the ant
    x += dx
    y += dy

    # check if the ant goes outside the boundary
    if y<0 or not(x+2 < y < x+4) or (x+1)**2 + (y-2)**2 < Rad**2 or (x+2)**2 + (y-1)**2 < Rad**2:
        # if so, bring it back where it was
        x -= dx
        y -= dy 
    # plot current position
    pl.plot([xp,x],[yp,y],c='r')
    
pl.show()

if nmoves >= Nmax:
    print('The ant is exhausted. We killed it.')
```
