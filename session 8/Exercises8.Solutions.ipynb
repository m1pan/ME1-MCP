```python
import copy
```

```python
class Student():
    ''' defines a student '''
    
    ''' attributes: cid, name list of marks, average'''
    
    def __init__(self, cid='', name='', marks=[]):
        self.cid = cid
        self.name = name
        self.marks = marks # this is a list of numbers
        
    def Average(self):
        # calculate the average of marks
        s = 0
        for value in self.marks:
            s += value
        Av = s / len(self.marks)
        return Av
    
    def Classify(self):
        # assign a degree classification to a student
        # average marks for this student
        av = self.Average()
        if av >= 70:
            classif = 'First'
        elif av >= 60:
            classif = '2.1'
        elif av >= 50:
            classif = '2.2'
        elif av >= 40:
            classif = 'Third'
        else:
            classif = 'Fail'
            
        return classif   
```

```python
class Registry(Student):
    ''' represent academic results of all students enrolled'''
    
    ''' attributes: dept, enrolments'''   
    
    def __init__(self, cid='', dept='', enrolments=0):
        self.dept = dept
        self.enrolments = enrolments
    
    def Statistics(self,classif):
        # determine how many students of the lot given have received the the classification classif
        count = 0
        for stud in self.enrolments:
            if stud.Classify() == classif:
                count += 1
        return count
```

```python
class Cohort(Student):
    ''' defines a cohort of student'''
    
    ''' attributes: year, list of students'''
    def __init__(self, year):
        self.year = year
        self.students= []

    def FindBestStudent(self):
        # find the student with the highest mark
        top = self.students[0].Average()
        best = self.students[0].name
        N = len(self.students)
        for i in range(1,N):
            if self.students[i].Average() > top:
                top = self.students[i].Average()
                best = self.students[i].name
        return best
          
    def FindDeansList(self):
        # create a deep copy of the cohort, to preserve the original from alias
        coh = copy.deepcopy(self)
        N = len(coh.students)
        # sort the students by mark, in descending order (make use of function sort)
        for i in range(N):
            for j in range(i+1,N):
                if coh.students[j].Average() > coh.students[i].Average():
                    (coh.students[i],coh.students[j]) = (coh.students[j],coh.students[i])
        
        # determine how many students will be taken into the Dean's list (10% of the cohort)
        Ndean = int(0.1*N)
        # form the list of names of top 10% students
        Dean = []
        for i in range(Ndean):
            Dean += [coh.students[i].name]
        return Dean
```

```python
for cohort in ME:

    best = cohort.FindBestStudent()
    print(cohort.year,best)
    
    dean = cohort.FindDeansList()
    print('Dean list: ',dean)
    
```
